```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Generate sample data
np.random.seed(42)
data = np.random.randint(1, 10,size=(10, 2)) 
data
```

```python
#Graph
x = data[:,0]
y = data[:,1]

# Plotting the data
plt.scatter(x, y, color='blue', marker='o')

# Adding labels to each point
for i, (x_coord, y_coord) in enumerate(zip(x, y)):
    plt.text(x_coord, y_coord, str(i), fontsize=12, ha='right', va='bottom')

plt.title('Scatter Plot with Numbered Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# Perform hierarchical clustering
Z = linkage(data, method='ward')  # Using Ward's method for linkage
```

```python
# Plot dendrogram
plt.figure(figsize=(8, 4))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

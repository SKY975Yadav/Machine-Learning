```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris['species'].unique()
iris
```

```python
iris.info()
```

# Parameters 

```python
# 1.Hue
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species') #hue = dependent var
```

```python
plt.figure(figsize=(12,8))
# sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species',palette='Reds')
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species',palette=['red','blue','green']) # cutom color
# palette='deep' (default)
# palette='bright'
# palette='muted'
# palette='coolwarm'
# palette='Blues', palette='Reds'
#https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,size='species')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,s=200,hue='species',palette=['red','blue','green'])
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,s=200,hue='species',palette=['red','blue','green'],alpha=0.8)
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,style='species',s=200,hue='species',palette=['red','blue','green'],alpha=0.8)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,style='species',hue='species',s=100)
plt.savefig('example_scatter.jpg')
```

```python
sns.displot(x='petal_length', data=iris, kde=True, rug=True)
```

```python
sns.histplot(data=iris,x='petal_length')
```

```python
sns.histplot(data=iris,x='petal_length',bins=30)
```

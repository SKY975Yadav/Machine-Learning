```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Create the pair plot
sns.pairplot(tips)

# Display the plot
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex')
```

```python
sns.pairplot(data=tips, hue='sex', palette='Set2')
```

```python
sns.pairplot(data=tips, vars=['total_bill', 'tip'])
```

```python
sns.pairplot(data=tips, x_vars=['total_bill', 'tip'], y_vars=['size'])
```

```python
sns.pairplot(data=tips, kind='kde')
#'scatter', 'kde', 'hist', 'reg'.
```

```python
sns.pairplot(data=tips, diag_kind='kde')
#'auto', 'hist', 'kde', None.
```

```python
sns.pairplot(data=tips, hue='sex', markers=['o', 's'])
```

```python
sns.pairplot(data=tips, height=1.5)
```

```python
sns.pairplot(data=tips, aspect=.8)
#Aspect Ratio= Width/Height
```

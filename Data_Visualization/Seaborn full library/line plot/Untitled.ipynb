```python
df1 = pd.read_excel('game.xlsx')
```

```python
df1
```

```python
# Using figure and seaborn
fig = plt.figure(figsize=(7, 5))
axes_upper = fig.add_axes([0.3, 0.3, 0.4, 0.4])  # left, bottom, width, height
sns.lineplot(x='Kills',y='Rank_Increase',data = df1)
axes_lower = fig.add_axes([0.8, 0.3, 0.4, 0.4])  # left, bottom, width, height
sns.lineplot(x='Damage',y='Rank_Increase',data = df1)
```

```python
# Create a figure with two subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 'axes' contains an array of subplots

# Plot on the first subplot (axes[0])
sns.lineplot(x='Kills', y='Rank_Increase', data=df1, ax=axes[0])
axes[0].set_title('Kills vs Rank Increase')

# Plot on the second subplot (axes[1])
sns.lineplot(x='Damage', y='Rank_Increase', data=df1, ax=axes[1])
axes[1].set_title('Damage vs Rank Increase')

# Show the plots
plt.tight_layout()
```

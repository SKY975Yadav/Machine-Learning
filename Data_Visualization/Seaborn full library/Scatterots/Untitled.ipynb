```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins").head(30)
df
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df,hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df,hue="sex",style = "sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df,hue="sex",style = "sex",size = "sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df,hue="sex",style = "sex",size = "sex",sizes =(100,70))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df,hue="sex",style = "sex",size = "sex",sizes =(100,70),palette = ["red","blue"]) # paette = Accent
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df,hue="sex",style = "sex",size = "sex",sizes =(100,70),palette = ["red","blue"],alpha = 0.5) # paette = Accent
plt.show()
```

```python
m = {"Male" : "*","Female" : "o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = df,hue="sex",style = "sex",size = "sex",sizes =(100,70),palette = ["red","blue"],markers= m) 
```

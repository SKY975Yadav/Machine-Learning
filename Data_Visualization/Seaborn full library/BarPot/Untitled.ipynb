```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample DataFrame
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [10, 15, 7, 20, 25, 10],
    'Group': ['X', 'X', 'X', 'Y', 'Y', 'Y']
})

data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Category', y='Value', data=data)
plt.title('Barplot with x and y Parameters')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', hue='Group', data=data)
plt.title('Barplot with Hue Parameter')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', data=data, order=['C', 'B', 'A'])
plt.title('Barplot with Ordered Categories')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', hue='Group', data=data, hue_order=['Y', 'X'])
plt.title('Barplot with Ordered Hue Categories')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', hue='Group', data=data, palette='pastel')
plt.title('Barplot with Custom Color Palette')
plt.show()
```

```python
sns.barplot(x='Value', y='Category', data=data, orient='h')
plt.title('Horizontal Barplot')
plt.show()
```

```python
import numpy as np

sns.barplot(x='Category', y='Value', data=data, estimator=np.sum)
plt.title('Barplot with Median Estimator')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', data=data, errorbar=('ci', 50), capsize=0.2)
plt.title('Barplot with Caps on Error Bars')
plt.show()
```

```python
# With dodge (default)
sns.barplot(x='Category', y='Value', hue='Group', data=data)
plt.title('Barplot with Dodge')
plt.show()

# Without dodge
sns.barplot(x='Category', y='Value', hue='Group', data=data, dodge=False)
plt.title('Barplot without Dodge')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', data=data)
plt.gca().patches[0].set_width(0.1)  # Adjust the width of bars manually
plt.title('Barplot with Custom Bar Width')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', data=data, linewidth=6)
plt.title('Barplot with Custom Line Width')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', data=data, edgecolor='black',linewidth=6)
plt.title('Barplot with Custom Edge Color')
plt.show()
```

```python
sns.barplot(x='Category', y='Value', data=data, errorbar=('ci', 95), err_kws={'color': 'purple','linewidth': 3})
plt.title('Barplot with Custom Error Bar Color')
plt.show()
```

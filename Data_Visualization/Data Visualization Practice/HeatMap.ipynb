```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from seaborn import load_dataset
flights = load_dataset('flights')
flights
```

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
flights_pivot
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="jet")

plt.title('Monthly Number of Passengers Over the Years')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

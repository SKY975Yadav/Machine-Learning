```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from seaborn import load_dataset
titanic = load_dataset('titanic')
titanic
```

```python
# Calculate survival rates by class and gender
survival_rates = titanic.groupby(['class', 'sex'])['survived'].mean().unstack()
# Plot the horizontal bar chart
survival_rates.plot(kind='barh', figsize=(10, 6))

plt.title('Survival Rates by Class and Gender')
plt.xlabel('Survival Rate')
plt.ylabel('Class')
plt.legend(title='Gender')
```

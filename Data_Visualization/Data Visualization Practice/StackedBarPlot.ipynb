```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from seaborn import load_dataset
titanic = load_dataset('titanic')
titanic
```

```python
survival_counts = titanic.groupby(['class', 'sex', 'survived']).size().unstack().fillna(0)
survival_counts
```

```python
survival_counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=['red', 'green'])

plt.title('Survival Counts by Class and Gender')
plt.xlabel('Class and Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(['Not Survived', 'Survived'], title='Survival Status')
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data = iris.data,columns = iris.feature_names)
df['Target'] = iris.target
df = df.head(100)
df
```

```python
x = df['sepal length (cm)']
y = df['sepal width (cm)']

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title('Linear Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(3.5,8.2)
plt.ylim(2.5,5.4)
plt.show()
```

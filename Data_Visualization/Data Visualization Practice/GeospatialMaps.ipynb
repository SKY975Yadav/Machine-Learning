```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
world = gpd.read_file("C:\\Users\\BAPS\\IdeaProjects\\Machine Learning Course\\Data_Visualization\\Data Visualization Practice\\110m_cultural\\ne_110m_admin_0_countries.shp")
world
```

```python
# Sample earthquake data
data = {
    'Longitude': [142.369, 84.726, 129.561, -123.456],
    'Latitude': [38.322, 28.652, -20.384, 46.755],
    'Magnitude': [7.0, 6.5, 5.8, 4.7]
}
```

```python
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['Longitude'], data['Latitude']))
gdf
```

```python
# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax, color='lightgrey')
gdf.plot(ax=ax, color='red', markersize=gdf['Magnitude'] * 10, alpha=0.5)

plt.title('Earthquake Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

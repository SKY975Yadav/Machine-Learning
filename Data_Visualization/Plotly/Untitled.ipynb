```python
import plotly.express as px

# Sample data
df = px.data.iris()
df
# # Create scatter plot
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species', title='Sepal Width vs Sepal Length')
fig.show()
```

```python
fig = px.scatter(df, x='sepal_width', y='sepal_length', size='sepal_length')
fig.show()
```

```python
fig = px.scatter(df, x='sepal_width', y='sepal_length', hover_data=['petal_width'])

fig.show()
```

```python
# Scatter plot with hover data
fig = px.scatter(df, x='gdpPercap', y='lifeExp', 
                 size='year', 
                 color='continent', 
                 hover_name='country',  # Country name when hovering
                 hover_data=['year', 'iso_alpha'],  # Additional data on hover
                 log_x=True, size_max=60)

fig.show()
```

```python

df = px.data.iris()
dffig = px.scatter(df, x='sepal_width', y='sepal_length', labels={'sepal_width': 'Sepal Width', 'sepal_length': 'Sepal Length'})
dffig
```

```python
fig = px.scatter(df, x='sepal_width', y='sepal_length', template='plotly_dark')
fig.show()
```

```python
df = px.data.gapminder()
fig = px.scatter(df, x='gdpPercap', y='lifeExp', animation_frame='year', animation_group='country', size='pop', color='continent', hover_name='country', log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
fig.show()
```

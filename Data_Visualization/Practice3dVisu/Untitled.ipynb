```python
import matplotlib.pyplot as plt
x_data = [1,4,6,8,12,15]
y_data = [2,3,5,5.6,7.2,8.3]
z_data = [3.3,4,6,8.1,8.3,11]
colors = [5,12,46,21,1,31]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# sc = ax.scatter(x_data, y_data, z_data,c = colors ,cmap='inferno',s=100)  # color_data is an optional fourth dimension
sc = ax.plot(x_data, y_data, z_data,c = 'red')  # color_data is an optional fourth dimension
# plt.colorbar(sc)  # Show color scale
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
import numpy as np
def some_function(x,y) :
    return x+y;

X, Y = np.meshgrid(x_data, y_data)
Z = some_function(X, Y)  # Calculate Z based on X and Y
print(X,Y,Z)
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.plot_wireframe(X, Y, Z, color='black')
plt.show()
```

```python
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
ax.bar3d(x_data, y_data, np.zeros(len(z_data)), dx=0.1, dy=0.1, dz=z_data, color='blue')
plt.show()
```
